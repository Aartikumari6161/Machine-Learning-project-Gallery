# outlier Remonig by min max scalor by IQR method
* Q1 
* Q3 
* IQR = Q3 - Q1

* min_range = q1 - (1.5*iqr)
* max_range = q3 + (1.5*iqr)

```python
q1 = df["net_margin"].quantile(0.25)
q3 = df["net_margin"].quantile(0.75)
```

```python
q1 , q3
```

```python
iqr = q3-q1
```

```python
iqr
```

```python
min_range = q1 - (1.5*iqr)
max_range = q3 + (1.5*iqr)
```

```python
min_range , max_range
```

```python
df2 = df[df["net_margin"]<max_range]
```
